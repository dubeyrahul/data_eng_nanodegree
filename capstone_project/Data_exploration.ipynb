### data description for i94 data:

- i94yr: 4 digit indicating which year the record is from
- i94mon: 4 digit indicating which month in numeric (for e.g. 4: April)
- i94cit: 3 digit code of origin and this digit is mapped to a country in SAS description, many cit are absent so ignore this
- i94res: 3 digit code of origin and this digit is mapped to a country in SAS description
- i94port: 3 letter code of destination city,State in US, mapping in SAS
- arrdate: is the Arrival Date in the USA. It is a SAS date numeric field that a 
   permament format has not been applied
   using this: https://stackoverflow.com/questions/36412864/convert-numeric-sas-date-to-datetime-in-pandas
- i94mode:There are missing values as well as not reported (9)
	1 = 'Air'
	2 = 'Sea'
	3 = 'Land'
	9 = 'Not reported'
- i94addr: US state (perhaps the destination state), there are some unknowns indicated with '99'
- depdate: Departure date from the US
- i94bir: Age of respondent in years
- i94visa: Visa codes into 3 categories
    1 = Business
    2 = Pleasure
    3 = Students
- count: summary stats
- dtadfile:

- biryear: 4 digit birth year
- dtaddto: Date admitted to US (allow to stay until?)

- admnum:
- fltno: Flight number
- visatype: Class of admission (H1, B1, etc.)

### Data cleaning

```python
i94port_map = dict()
with open('i94port.txt') as f:
    for line in f.readlines():
        line = line.strip().split("=")
        code = eval(line[0]).strip()
        location = eval(line[1]).strip()
        i94port_map[code] = location
```

```python
i94res_map = dict()
with open('i94res.txt') as f:
    for line in f.readlines():
        line = line.strip().split("=")
        code = eval(line[0])
        location = eval(line[1]).strip()
        i94res_map[code] = location
```

```python
i94addr_map = dict()
with open('i94addr.txt') as f:
    for line in f.readlines():
        line = line.strip().split("=")
        code = eval(line[0]).strip()
        location = eval(line[1]).strip()
        i94addr_map[code] = location
```
